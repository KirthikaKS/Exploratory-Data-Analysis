```python
sns.histplot(df["2023 population"],bins = 10, kde = True)
plt.title("Distribution of 2023 population")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby("continent")["2023 population"].mean()
```

```python
df.groupby("continent")["2023 population"].mean().max()
```

```python
df.groupby("continent")["2023 population"].mean().min()
```

```python
df.groupby("continent")["2023 population"].mean().sort_values(ascending = False).tail(5)
```

```python
top_10_countries = df.sort_values(by = "2023 population", ascending = False).head(10)
```

```python
top_10_countries
```

```python
sns.barplot(x="2023 population",y = "country", data = top_10_countries)
plt.title("Top 10 countries by population in 2023")
plt.xlabel("Population")
plt.ylabel("Country")
plt.show()
```

```python
df.rename(columns = {"density (km²)": "density (km2)"},inplace = True)
```

```python
sns.scatterplot(data = df,x="2023 population", y="density (km2)")
plt.title("Population and Density scatter plot")
plt.xlabel("2023 population")
plt.ylabel("Density")
plt.show()
```

```python
df.rename(columns = {"area (km²)": "area (km2)"},inplace = True)
```

```python
df.groupby("continent")["area (km2)"].sum().sort_values(ascending = False)
```

```python
sns.violinplot(x="continent",y="density (km2)", data=df, palette = "viridis")
plt.title("Population Density Distribution Across Continents")
plt.xlabel("Continent")
plt.ylabel("population density")
plt.show()
```

```python
sns.boxplot(x="continent",y="density (km2)",data =df, palette ="muted", showfliers = False)
plt.title("Population Density Distribution Across Continents (Without Outlier)")
plt.xlabel("Continent")
plt.ylabel("population density")
plt.show()
```
