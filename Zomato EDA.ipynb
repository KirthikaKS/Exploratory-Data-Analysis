```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
#joins the two dataframes , on tell on which column the join operation has to be performed and how tells whic join has to be performed, since our main focus is df we use left join
final_df = pd.merge(df,df_country,on='Country Code',how ='left')
```

```python
final_df.head()
```

```python
#To check data types
final_df.dtypes
```

```python
# Finding in which country many transactions happen
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_val = final_df.Country.value_counts().values
```

```python
country_val
```

```python
#Pie chart - finding which country has max transactions - just displaying top 3
plt.pie(country_val[:3], labels = country_names[:3],autopct = "%1.2f%%")
```

Observation

1 - Zomato's maximum transactions are in India followed by United States and United Kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating count'})
```

```python
ratings
```

Obseration 

When the rating is between 4.5 -  4.9 --> Excellent

When the rating is between 4.0 -  4.9 --> Very Good

When the rating is between 3.5 -  3.9 --> Good

When the rating is between 3.0 -  3.4 --> Average

When the rating is between 2.5 -  2.9 --> Poor

When the rating is between 0.0 --> No rating has been provided


```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = "Aggregate rating", y = "Rating count", data = ratings)
```

```python
#matching rating colors to bars

import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = "Aggregate rating", y = "Rating count", data = ratings,hue='Rating color',palette = ['blue','red','orange','yellow','green','darkgreen'])
```

Observation

1) Not rated count is very high

2) Maximun no. of ratings is between 2.5 to 3.4

```python
## Count plot - how frequently the color is present 
sns.countplot(x='Rating color', data = ratings, palette = ['blue','red','orange','yellow','green','darkgreen'])
```
